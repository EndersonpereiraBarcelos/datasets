```python
import pandas as pd 

football = pd.read_csv('./dados/football.csv')

football.head(50)
```

```python
footballSerieA.to_csv('./dados/footballSerieA.csv', index=False)
footballSerieA.describe()
```

```python
import pandas as pd
import sqlite3

footballSerieA = pd.read_csv('./dados/footballSerieA.csv')

conn = sqlite3.connect('soccer.db')

footballSerieA.to_sql('footballSerieA', conn, if_exists='replace', index=False)

print("Tabela 'footballSerieA' criada com sucesso no banco de dados 'soccer.db'.")

conn.close()
```
