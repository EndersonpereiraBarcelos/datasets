```python
import pandas as pd
import sqlite3

LgLaLiga = pd.read_csv('./dados/lglaliga.csv')

conn = sqlite3.connect('soccer.db')

LgLaLiga.to_sql('lglaliga', conn, if_exists='replace')

print('Dados Inseridos')

conn.close()

```
